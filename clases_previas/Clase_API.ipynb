# Ejercicio 1: Obtener redes de bicicletas disponibles en citybik y obtener información de las estaciones de una de las redes

```python
def get_networks():
  url = 'https://api.citybik.es/v2/networks'
  response = requests.get(url)

  if response.status_code == 200:

    data = response.json()

    return data.get("networks")
data = get_networks()
```

```python
def get_stations(network_id):
  url = f'https://api.citybik.es/v2/networks/{network_id}'
  response = requests.get(url)
  if response.status_code == 200:
    data = response.json()
    if 'stations' in data['network'] and isinstance(data['network']['stations'], list):
      return data['network']['stations']
    else:
      return []
network_id = 'sevici'
estaciones = get_stations(network_id)
estaciones
```

```python
# Recorremos la lista de estaciones
for dicc in estaciones:
  # Acceder a la dirección de la estación
  direccion = dicc['extra']['address']
  # Acceder al númer de bicicletas disponibles
  free_bikes = dicc['free_bikes']
  # Mostrar por pantalla un mensaje que diga "La estación en la calle XXX tiene disponible XXX bicicletas."
  print(f"La estación en la calle {direccion} tiene disponible {free_bikes} bicicletas.")
```

# Ejercicio 4: RapidAPI

```python
rapidapi_key  = os.getenv('x-rapidapi-key')
```

```python
def get_raw_data():
  # Identificadores de la API
  url = "https://idealista7.p.rapidapi.com/listhomes"

  querystring = {"order":"relevance","operation":"sale","locationId":"0-EU-ES-28-07-001-079","locationName":"Madrid","numPage":"1","maxItems":"40","location":"es","locale":"es"}

  headers = {
    "x-rapidapi-key": rapidapi_key,
    "x-rapidapi-host": "idealista7.p.rapidapi.com"
  }
  # Realizamos la primera búsqueda
  return requests.get(url, headers=headers, params=querystring).json()
```
