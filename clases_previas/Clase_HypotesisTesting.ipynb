```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import stats
import statsmodels.api as sm
```

Hipótesis Nula (H_0): Media de botellas igual a 500ml

Hipótesis Alternativa (H_1): Media de botellas distinta a 500ml

```python
mu_0 = 500 # Valor bajo la hipótesis nula (valor promedio que afirma el fabricante)

media = 495

desviacion = 15

n = 40

error_estandar = desviacion/n**0.5
error_estandar
```

```python
diferencia = media - mu_0
diferencia
```

```python
estadistico = diferencia/error_estandar
estadistico
```

```python
p_valor = 2*norm.cdf(estadistico)
p_valor
```

```python
alpha = 0.01

print(f"p-valor = {p_valor:.6f}")

if p_valor < alpha:
    print("Rechazamos H0: hay evidencia contra la afirmación del fabricante.")
    print("La diferencia no parece deberse al azar.")
else:
    print("No rechazamos H0: la muestra es compatible con la afirmación del fabricante.")
    print("La diferencia observada podría deberse al azar.")
```

```python
n_casas = 1000
np.random.seed(0) # Para reproducir el experimento
X = np.random.rand(n_casas, 1) # Generamos la variable la que depende el precio de la vivienda
y = 6 + 10 * X + np.random.randn(n_casas, 1)

X = sm.add_constant(X)

# Crear un modelo de regresión lineal
modelo = sm.OLS(y, X)

# Ajustar el modelo a los datos
resultados = modelo.fit()

# Obtener el resumen de la regresión
print(resultados.summary())
```

```python
p_valor = resultados.pvalues[1]
p_valor
```

```python
# Realizar la prueba de hipótesis
if p_valor < alpha:
    print("Rechazamos la hipótesis nula.")
    print("Hay evidencia estadística significativa de que el coeficiente calculado es distinto de 0.")
else:
    print("No podemos rechazar la hipótesis nula.")
    print("No hay evidencia estadística suficiente para afirmar que el coeficiente calculado sea distinto de 0.")
```

# Calcular si los niveles de colesterol entre un grupo que ha recibido un tratamiento y otro control son distintos estadísticamente hablando.

```python
grupo_sin_tratamiento = [180, 190, 200, 195, 185, 210, 205, 188, 192, 198]
grupo_con_tratamiento = [160, 165, 155, 170, 175, 180, 168, 172, 162, 178]
```

```python
# Primer paso es comprobar q los datos de ambos grupos siguen una distribución normal
print(stats.normaltest(grupo_sin_tratamiento))
print(stats.normaltest(grupo_con_tratamiento))
```

```python
# La varianza de los grupos debe ser la misma
print(stats.levene(grupo_sin_tratamiento, grupo_con_tratamiento))
```

```python
# Realizar el contraste de hipótesis
t_stat, p_valor = stats.ttest_ind(grupo_sin_tratamiento, grupo_con_tratamiento)

# Definir el nivel alpha (0.05 es comúnmente utilizado)
nivel_alpha = 0.05

print("Estadística t:", t_stat)
print("Valor p:", p_valor)

# Realizar la prueba de hipótesis
if p_valor < nivel_alpha:
    print("Rechazamos la hipótesis nula.")
    print("Hay evidencia estadística significativa de que existe una diferencia en los niveles de colesterol entre los grupos.")
else:
    print("No podemos rechazar la hipótesis nula.")
    print("No hay evidencia estadística suficiente para afirmar que exista una diferencia en los niveles de colesterol entre los grupos.")
```

En caso de que queramos comprobar esa diferencia entre las medias, pero no podamos utilizar el t-test debido a que no cumplimos el requerimiento de normalidad, podemos utilizar tests no-paramétricos:

- Mann Whitney U: scipy.stats.mannwhitneyu
- Wilcoxon: scipy.stats.wilcoxon
