```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
```

```python
url = "https://quotes.toscrape.com/"

# Realizar una petición al sitio web
response = requests.get(url)

# Comprobamos si hemos tenido éxito en la petición
if response.status_code == 200:
  # En caso de tener éxito le damos el contenido de la página a la librería BeautifulSoup
  soup = BeautifulSoup(response.content)
  # Lista de etiquetas div
  ls_div = soup.find_all('div', class_="quote")
  # Recorremos la lista de div
  for etiqueta in ls_div:
    # Extraemos el texto
    texto = etiqueta.find('span',class_="text").get_text()
    autor = etiqueta.find('small',class_="author").get_text()
    print(f'Texto:{texto}')
    print(f'Autor:{autor}')
    print('-'*50)
```

# Craiglist

```python
data = []

# URL de la página de búsqueda en Craigslist (puedes cambiar 'sfbay' a la región que desees)
url = 'https://sfbay.craigslist.org/search/sss?query=laptop'

# Headers para simular un navegador real
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

response = requests.get(url, headers=headers)

# Comprobamos si hemos tenido éxito en la petición
if response.status_code == 200:
  # En caso de tener éxito le damos el contenido de la página a la librería BeautifulSoup
  soup = BeautifulSoup(response.content)

  # Identificar todos los elementos de anuncios
  anuncios = soup.find_all('li', class_='cl-static-search-result')

  for ad in anuncios:
    titulo = ad.find('div',class_='title')
    precio = ad.find('div',class_='price')
    localizacion = ad.find('div',class_='location')
    if all(elem is not None for elem in [titulo, precio, localizacion]):
      titulo = titulo.get_text()
      precio = float(precio.get_text()[1:].replace(",",""))#$1,100
      localizacion = localizacion.get_text().strip()
      # Almacenamos los datos
      data.append((titulo, precio,localizacion))
```

```python
df = pd.DataFrame(data, columns=['Titulo','Precio','Localización'])
df
```

# Imágenes

```python
def scrape_images(url, ruta):
  response = requests.get(url)

  # Comprobamos si hemos tenido éxito en la petición
  if response.status_code == 200:
    # Crear una instancia de BeautifulSoup para analizar la página
    soup = BeautifulSoup(response.text)

    # Obtener todas las etiquetas de imagen
    ls_imgs = soup.find_all('img')

    # Iteramos sobre la lista de etiquetas de imágenes
    for i,img_tag in enumerate(ls_imgs):
      # Acceder a la url de la imagen
      src = img_tag.get('src')
      # si efectivamente se ha encontrado la url
      if src:
        # Descargarla en la ruta correspondiente
        urllib.request.urlretrieve(src, f'{ruta}/imagen_{i}.jpg')


url = 'https://www.pinterest.es/maitebortrodenas/imagenes-de-valencia/'

scrape_images(url,'data')
```

```python
async def extract_linkedin_content(urls):
    data = []
    async with async_playwright() as p:
      # Abrimos el navegador
      browser = await p.chromium.launch(headless=True)
      # Abrimos la pestaña
      page = await browser.new_page()

      for url in tqdm(urls):
        try:
          await page.goto(url, timeout=30000)
          text_content = await page.locator('body').inner_text()
        except Exception as e:
          print(f"Excepción del tipo {e} con la url: {url}")
        else:
          data.append({'url': url, 'texto': text_content})
          print(f"Texto extraído de la url: {url}")
      browser.close()

      return pd.DataFrame(data)


df_linkedin = await extract_linkedin_content(ls_urs)
```
