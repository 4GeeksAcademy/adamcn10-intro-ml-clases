```python
shopping_list = [[10.5, 3.25, 5.0], [2.99, 7.0], [1.5, 4.75, 9.99]]

def get_total_amount(lista):
  # Inicializamos un contador
  total = 0
  # Iteramos sobre la lista
  for sublist in lista:
    # Iteramos sobre las sublistas
    for importe in sublist:
      # Agregar el importe al contador
      total += importe
  # Devuelvo el contador
  return total

get_total_amount(lista=shopping_list)
```

# Acronimos

Hola mundo -> HM

```python
frase = "Hola mundo"

def get_acronym(frase):
  # Inicializamos la variable respuesta ""
  respuesta = ""
  # Rompemos por espacios con split
  palabras = frase.split()
  # Iteramos sobre las palabras de la frase
  for palabra in palabras:
    # Accedemos al indice 0 de cada palabra la convertimos a mayúscula
    inicial = palabra[0].upper()
    # La almacenamos en la variable respuesta
    respuesta += inicial
  # Devolvemos la variable respuesta
  return respuesta
get_acronym(frase)
```

# map, filter, sort

```python
personas = [
    {"nombre": "Alice", "edad": 25, "is_overage":True},
    {"nombre": "Bob", "edad": 13},
    {"nombre": "Charlie", "edad": 22},
    {"nombre": "David", "edad": 15},
    {"nombre": "Eve", "edad": 28}
]
```

```python
personas_1 = list(map(lambda x: {"nombre": x['nombre'], "edad": x['edad'], "is_overage":x['edad']>=18}, personas))
personas_1
```

```python
personas_overage = list(filter(lambda x:x['is_overage']==True, personas_1))
personas_overage
```

```python
# Por edad de menos a mas
personas_overage.sort(key=lambda x: x['edad'], reverse=False)
personas_overage
```

```python
n = 10**7  # Número de filas
categorias = ['A', 'B', 'C', 'D']
numeros = np.random.randn(n)

cat_column = np.random.choice(categorias, size=n)
df_pandas = pd.DataFrame({'categoria': cat_column, 'valor': numeros})
```
